The Python script below shows the way to plot a line in 3D.

```python
%matplotlib notebook
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection="3d")

z_line = np.arange(0, 15, 0.001)
x_line = np.log(z_line + 1)
y_line = np.power(z_line - 5,3)
ax.plot3D(x_line, y_line, z_line)

plt.show()
```

Implement 3D harmonographs for the frequency ratios in the major chord (4:5:6) and the minor chord (10:12:15). You should able to see 2D harmonograph in different projections.

```python
# major chord
f = [4,5,6]
p = [0,0,0]

# minor chord
#f = [10,12,15]
#p = [0,0,0]

fig = plt.figure()
ax = plt.axes(projection="3d")

t = np.arange(0, 15, 0.001)
z_line = np.sin(2*np.pi * f[0] * t + p[0])
x_line = np.sin(2*np.pi * f[1] * t + p[1])
y_line = np.sin(2*np.pi * f[2] * t + p[2])
ax.plot3D(x_line, y_line, z_line)

plt.show()
```
