```python
fig = plt.figure()
ax = plt.axes(projection="3d")

t = np.linspace(0, 15, 1000)
z_line = np.cos(t)
x_line = np.cos(t)
y_line = np.sin(t)
ax.plot3D(x_line, y_line, z_line, 'gray')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
def z_function(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
