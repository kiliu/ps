A simple melody player is implemented in "Introduction" as follows:

```python
import sys
sys.path.append('../ThinkDsp')
import thinkdsp
```

```python
key = 67
note_list = [4,4,5,7,7,5,4,2,0,0,2,4,4,2,2]
wave = thinkdsp.rest(0)
for note in note_list:
    midi_number = key + note
    w = thinkdsp.make_note(midi_number, 0.5)
    w.apodize()
    wave = wave | w
wave.make_audio()
```

Modify the above implementation so that the melody player supports note values (the durations of notes). The necessary parameters are:

```python
key = 67
note_list = [7, 0, 2, 4, 5, 7, 0, 0]
note_value = [1, 1/2, 1/2, 1/2, 1/2, 1, 1, 1]
bpm = 120
```

Write your answer in the cell below.

```python
beat_length = 60/bpm
l = len(note_list)
wave = thinkdsp.rest(0)
for i in range(l):
    midi = note_list[i] + key
    w = thinkdsp.make_note(midi,note_value[i]*beat_length)
    w.apodize()
    wave = wave | w
wave.make_audio()
```
