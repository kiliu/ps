# Noise

```python
%matplotlib inline
import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot

import matplotlib.pyplot as plt
```

## Uncorrelated noise

```python
signal = thinkdsp.UncorrelatedUniformNoise()
uu_noise = signal.make_wave(duration=2, framerate=11025)
uu_noise.make_audio()
```

### Waveform

```python
seg = uu_noise.segment(1,0.01)
seg.plot(linewidth=1, alpha=0.5)
```

### Spectrum

```python
seg = uu_noise.segment(1,0.1)
spectrum = seg.make_spectrum()
spectrum.plot(linewidth=1, alpha=0.5)
```

### Power spectrum

```python
spectrum.plot_power(linewidth=1, alpha=0.5)
```

### Power spectrum in log scale

```python
plt.xscale("log")
plt.yscale("log")
spectrum.plot_power(linewidth=1, alpha=0.5)
```

### Integrated spectrum

```python
int_spec = spectrum.make_integrated_spectrum()
int_spec.plot_power(alpha=0.5)
```
