# Lab 04a: Decoding wav file

```python
%matplotlib inline

import binascii
```

```python
with open("cello.wav", "rb") as f:
    print("RIFF chunk")
    byte = f.read(1)
    for i in range(12):
        print(i, byte)
        byte = f.read(1)
    print()
    print("FORMAT chunk")
    for i in range(24):
        print(i, byte)
        byte = f.read(1)   
    print()    
    print("DATA chunk")
    for i in range(50):
        print(i, byte)
        byte = f.read(1) 
```
