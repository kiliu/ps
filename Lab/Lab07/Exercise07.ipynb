Implement just intonation and <a href="https://en.wikipedia.org/wiki/Quarter-comma_meantone">meantone temperament</a> for <i>SoundSyn</i>. Test whether you can  identify the differences among four tunings.

```python
import sys
sys.path.append('../SoundSyn')
import soundSyn
import numpy as np
```

```python
def jiScale(start_index):
    x = [1, 16/15, 9/8, 6/5, 5/4, 4/3, 45/32, 3/2, 8/5, 5/3, 9/5, 15/8]
    
    fp = [0]*89
    fp[0] = 0
    freq = soundSyn.etScale[start_index]
    
    for i in range(12):
        fp[start_index+i] = freq * x[i]
        
    for p in range(5):
        for q in range(12):
            fp[start_index + q + (p+1)*12] = fp[start_index + q] * 2**(p+1)
    return(fp)

jiScale =  jiScale(12) # G1
```

```python
ssJI = soundSyn.SoundSyn()
ssJI.set_scale(jiScale)
ssJI.set_file("bwvAnhII114.txt")
wave = ssJI.make_wave()
wave.make_audio()
```

```python
def mtScale(start_index):
    d = 5**(1/4)
    x = [1, (d**(-5))*8, np.sqrt(5)/2, (d**(-3))*4, 5/4, 2/(5**(1/4)), (d**6)/8, 5**(1/4), 8/5, 5**(3/4)/2, (d**(-2))*4, 5**(5/4)/4]
    
    fp = [0]*89
    fp[0] = 0
    freq = soundSyn.etScale[start_index]
    
    for i in range(12):
        fp[start_index+i] = freq * x[i]
        
    for p in range(5):
        for q in range(12):
            fp[start_index + q + (p+1)*12] = fp[start_index + q] * 2**(p+1)
    return(fp)

mtScale =  mtScale(12) # G1
```

```python
ssMT = soundSyn.SoundSyn()
ssMT.set_scale(mtScale)
ssMT.set_file("bwvAnhII114.txt")
wave = ssMT.make_wave()
wave.make_audio()
```
